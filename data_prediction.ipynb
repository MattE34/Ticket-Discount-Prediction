### Predicting on Testing Data

```python
test_df['Departure Date'] = pd.to_datetime(test_df["Departure Date"], format="%m/%d/%Y")
test_df['Departure Date']
```

```python
test_df['Departure Day'] = test_df['Departure Date'].dt.day
test_df['Departure Month'] = test_df['Departure Date'].dt.month
test_df['Departure Year'] = test_df['Departure Date'].dt.year
test_df
```

```python
test_df = test_df.merge(country_code_map_df, on="Airport Country Code", how="left")
test_df
```

```python
def discount(row):
    if row['Age'] < 5 or row['Age'] > 85:
        return 'Yes'
    elif row['Departure Month'] == 1 or row['Departure Month'] == 12:
        return 'Yes'
    elif row['Gender'] == 'Female' and row['Ticket Price'] % 4 == 0:
        return 'Yes'
    elif row['Nationality'] == row['Country Name']:
        return 'Yes'
    else:
        return 'No'
    


test_df['Eligible_For_Discount'] = test_df.apply(discount, axis=1)
test_df
```
